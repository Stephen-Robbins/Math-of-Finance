#### Solution

Given $\Delta$ of a European put option is $\Delta = N(d_1) - 1$, and we have $\frac{\partial P_E(S, t)}{\partial S} = -0.5$. Thus, we equate the given delta to $-0.5$:

$$
N(d_1) - 1 = -0.5
$$

Solving for $N(d_1)$ gives:

$$
N(d_1) = 0.5
$$

Since the cumulative normal distribution function $N(d_1)$ is symmetric about 0, and $N(0) = 0.5$, we deduce that $d_1 = 0$.

#### Vega ($\nu=\frac{\partial V}{\partial \sigma}$)
 For a European option, Vega is given by:

$$
\nu = S \sqrt{T-t} N'(d_1)
$$

Given $d_1 = 0$ and knowing $N'(0) = \frac{1}{\sqrt{2\pi}}$ (since $N'(x)$ is the probability density function of a standard normal distribution), we substitute $d_1 = 0$ into the Vega formula:

$$
\nu = S \sqrt{T-t} \frac{1}{\sqrt{2\pi}}
$$

Given the stock price $S = 100$ and time until expiration $T-t = 4$ years, the Vega becomes:

$$
\nu = 100 \sqrt{4} \frac{1}{\sqrt{2\pi}}
$$

Simplifying yields:

$$
\nu =  \frac{200}{\sqrt{2\pi}}
$$

Consider an at-the-money European call option on this stock with 1 year to expiration. What is the most likely value of the stock price at expiration? Please round your numerical answer to 2 decimal places.


### Solution to Question 8

Given the stock price $S(t)$ follows the Geometric Brownian Motion described by:

$$
dS = \mu S dt + \sigma S dW,
$$

where $\mu = 0.13$, $\sigma = 0.20$, and the current stock price is $S(0) = 100$. For an at-the-money European call option with 1 year to expiration, we are interested in finding the most likely value of the stock price at expiration.

#### Log-Normal Distribution Parameters

For a Geometric Brownian Motion, the stock price at time $T$ has a log-normal distribution. The parameters of this distribution are given by:

$$
\mu_1 = \ln[S(0)] + \left(\mu - \frac{1}{2}\sigma^2\right)(T - 0)
$$

and

$$
\sigma_1^2 = \sigma^2(T - 0)
$$

Substituting the given values, $\mu = 0.13$, $\sigma = 0.20$, $S(0) = 100$, and $T = 1$ year, we get:

$$
\mu_1 = \ln[100] + \left(0.13 - \frac{1}{2}(0.20)^2\right)(1)
$$

$$
\sigma_1^2 = (0.20)^2(1)
$$

#### Most Likely Value at Expiration

The mode of the log-normal distribution, which represents the most likely value of the stock price at expiration, is given by:

$$
x = e^{\mu_1 - \sigma_1^2}
$$

Substituting $\mu_1$ and $\sigma_1^2$ into this formula gives:

$$
x = e^{\ln[100] + \left(0.13 - \frac{1}{2}(0.20)^2\right)(1) - (0.20)^2(1)}\approx 107.25
$$

### Question 10

Consider now at time $t$ a portfolio $\Pi$ that holds one long European call option, one short European put option, $\alpha$ units of the stock $S(t)$, and an amount equal to $2Ee^{-r(T-t)}$ dollars in the bank that continuously compounds at interest rate $r$. Mathematically we define this as follows:

$$
\Pi = C_E(S, t) - P_E(S, t) + S(t)\alpha + 2Ee^{-r(T-t)}
$$

Suppose we compute $\alpha$ to be such that the value of the portfolio $\Pi$ will be unaffected by small changes in the underlying stock price $S$. With $\alpha$ set to this value, what is the value of this portfolio at expiration if $E = \$100$? Please round your numerical answer to the nearest integer dollar.


#### Solution to Question 10

Given the portfolio:

$$
\Pi = C_E(S, t) - P_E(S, t) + S(t)\alpha + 2Ee^{-r(T-t)}
$$

We want to find the value of $\alpha$ such that the portfolio is delta-neutral, meaning that the portfolio's value is unaffected by small changes in $S$. The deltas of the call and put options are $N(d_1)$ and $N(d_1)-1$ respectively. To make the portfolio delta-neutral, we set the total delta to zero:

$$
\frac{\partial \Pi}{\partial S} = N(d_1) - (N(d_1) - 1) + \alpha = 0
$$

Solving for $\alpha$ gives:

$$
\alpha  = -1
$$

Now that we have $\alpha = -1$, we can use put-call parity to find the value of the portfolio at expiration. Put-call parity is given by:

$$
C_E(S, T) - P_E(S, T) = S(T) - Ee^{-r(T-t)}
$$

Substitute $\alpha$ and rearrange the terms to find the value of the portfolio $\Pi$ at expiration:

$$
\Pi = C_E(S, T) - P_E(S, T) - S(T) + 2Ee^{-r(T-t)}
$$

Using put-call parity, we replace $C_E(S, T) - P_E(S, T)$ with $S(T) - Ee^{-r(T-t)}$:

$$
\Pi = (S(T) - Ee^{-r(T-t)}) - S(T) + 2Ee^{-r(T-t)}
$$

Simplifying the above expression:

$$
\Pi = Ee^{-r(T-t)}
$$

So at expiration 
$$
\Pi = E.
$$


# Volatility Smile

## Implied Volatility

The implied volatility $\sigma_{imp}$ is calculated by inverting the Black-Scholes formula:

$$
\sigma_{imp} = BS^{-1}(S_0, K, r, T, C)
$$

where $S_0$ is the current asset price, $K$ is the strike price, $r$ is the risk-free rate, $T$ is the time to expiration, and $C$ is the market price of the option.

## Volatility Smile Pattern

The volatility smile refers to the observed pattern of implied volatilities $\sigma_{imp}$ across different strike prices $K$ for options with the same underlying asset and expiration date $T$.

$$
\sigma_{imp} = f(K)
$$

This pattern forms a smile-shaped curve when plotted against the strike price.

## Volatility Behavior

Empirical observations suggest that volatility exhibits certain behaviors:

### Mean Reversion

Volatility tends to revert to a long-term mean level over time. This can be modeled using a mean-reverting process:

$$
d\sigma_t = \kappa(\theta - \sigma_t)dt + \xi dW_t
$$

where $\kappa$ is the mean-reversion speed, $\theta$ is the long-term mean volatility, and $\xi$ is the volatility of volatility.

### Volatility Clustering

Periods of high (low) volatility tend to be followed by periods of high (low) volatility, exhibiting momentum or clustering behavior.

### Leverage Effect

For equity options, a decrease in the underlying stock price tends to increase its leverage and volatility, contributing to the volatility smile pattern.

```python
import numpy as np
import scipy.stats as stats
from scipy.optimize import brentq
import matplotlib.pyplot as plt
import pandas as pd

# Define the Black-Scholes price for a European call option
def black_scholes_call(S0, E, T, r, sigma):
    d1 = (np.log(S0 / E) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return S0 * stats.norm.cdf(d1) - E * np.exp(-r * T) * stats.norm.cdf(d2)

# Function to calculate implied volatility
def implied_volatility_call(S0, E, T, r, market_price):
    def difference_in_price(sigma):
        return black_scholes_call(S0, E, T, r, sigma) - market_price
    implied_vol = brentq(difference_in_price, 0.01, 3)
    return implied_vol

# Market data
S0 = 189
r = 0.0531
T = (pd.to_datetime('2025-01-17') - pd.to_datetime('2024-02-07')).days / 365.25

# Options data
options_data = {
    'Strike Price': [165, 170, 175, 180, 185, 190, 195, 200, 205, 210],
    'Current Price': [36.56, 32.97, 29.73, 25.75, 22.70, 19.86, 17.00, 14.50, 12.61, 10.23]
}

# Calculate implied volatility for each option
implied_vols = []
for E, market_price in zip(options_data['Strike Price'], options_data['Current Price']):
    iv = implied_volatility_call(S0, E, T, r, market_price)
    implied_vols.append(iv)

# Plot the implied volatility smile
plt.figure(figsize=(10, 5))
plt.plot(options_data['Strike Price'], implied_vols, marker='o')
plt.xlabel('Strike Price')
plt.ylabel('Implied Volatility')
plt.title('Implied Volatility Smile for AAPL Options')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.stats import norm
from scipy.optimize import newton

# Black-Scholes formula for European call option
def call_option_price(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    return call_price

# Black-Scholes formula for European put option
def put_option_price(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    put_price = K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
    return put_price

# Objective function to find the root (implied volatility)
def implied_volatility_objective(sigma, S, K, T, r, market_price, option_type='call'):
    if option_type == 'call':
        return call_option_price(S, K, T, r, sigma) - market_price
    elif option_type == 'put':
        return put_option_price(S, K, T, r, sigma) - market_price

# Function to calculate implied volatility
def calculate_implied_volatility(S, K, T, r, market_price, option_type='call'):
    # Initial guess for volatility
    sigma_initial_guess = 0.2
    # Use Newton-Raphson method to find the root of the objective function
    implied_vol = newton(implied_volatility_objective, sigma_initial_guess, args=(S, K, T, r, market_price, option_type))
    return implied_vol

# Example usage:
S = 100  # Current stock price
K = 100  # Strike price
T = 1    # Time to expiration in years
r = 0.05 # Risk-free interest rate
market_price = 10  # Market price of the option

# Calculate implied volatility for a call option
implied_vol_call = calculate_implied_volatility(S, K, T, r, market_price, option_type='call')
print(f"Implied Volatility for Call Option: {implied_vol_call:.6f}")

# Calculate implied volatility for a put option
implied_vol_put = calculate_implied_volatility(S, K, T, r, market_price, option_type='put')
print(f"Implied Volatility for Put Option: {implied_vol_put:.6f}")
```

### Volatility Harvesting Strategies

#### Straddle/Strangle

$$
\begin{aligned}
\text{Straddle Payoff} &= \max(S_T - K, 0) + \max(K - S_T, 0) \\
\text{Strangle Payoff} &= \max(S_T - K_C, 0) + \max(K_P - S_T, 0)
\end{aligned}
$$

Straddles and strangles involve buying a call and a put option (straddle) or a higher strike call and a lower strike put (strangle). These strategies profit from high realized volatility, as the options become more valuable when the underlying asset price moves significantly in either direction.

#### Volatility Spread

$$
\begin{aligned}
\text{Long Volatility Spread} &= \nu_C(K_1, \sigma_1) - \nu_C(K_2, \sigma_2) \\
\text{Short Volatility Spread} &= \nu_P(K_1, \sigma_1) - \nu_P(K_2, \sigma_2)
\end{aligned}
$$

Volatility spreads involve taking offsetting positions in options with different strike prices and/or maturities, exploiting the difference in their vega exposures. Long volatility spreads profit from volatility increases, while short volatility spreads benefit from volatility decreases.

#### Variance Swap

$$
\text{Variance Swap Payoff} = N(\text{Realized Variance} - \text{Strike Variance})
$$

Variance swaps are over-the-counter derivatives that allow traders to speculate on the future realized variance of an underlying asset. The payoff is based on the difference between the realized variance and the strike variance, enabling traders to profit from volatility mispricing.

#### Volatility Arbitrage

$$
\begin{aligned}
\sigma_{\text{implied}} &\neq \sigma_{\text{expected}} \\
\text{Arbitrage Profit} &= \text{Option Premium} - \text{Hedging Cost}
\end{aligned}
$$

Volatility arbitrage involves identifying and exploiting discrepancies between the implied volatility from option prices and the expected future volatility. Traders can construct delta-neutral portfolios to capture the difference between the option premium and the cost of hedging, profiting from volatility mispricing.



### Problem 1

As a United States Senator in early 2020, you have confidential information about an impending pandemic that is likely to cause significant market volatility. To capitalize on the expected increase in market volatility without directly buying put optionswhich could be construed as insider tradingyou decide to employ a strategy that involves buying and selling call options on the S&P 500 Index  to create a delta-neutral position.

#### Background Information:

- The S&P 500 Index is currently at 3,200 points.
- You are considering 6-month European call options on the S&P 500 with a strike prices rainging between 3000-3500 in increments of 100.
- The call options are trading with an implied volatility of 25%.
- The continuously compounded risk-free interest rate is 1% per annum.

You can only buy/sell 100 of each option. What is the maximum profit you can achieve while maintaining a delta neutral portfolio.


```python
from scipy.optimize import linprog
import numpy as np
from scipy.stats import norm

# Given parameters
S0 = 3200  # Current S&P 500 Index level
Ks = np.array([3000, 3100, 3200, 3300, 3400, 3500])  # Strike prices
r = 0.01  # Risk-free interest rate
T = 0.5  # Time to expiration in years
sigma = 0.25  # Implied volatility (25%)


# Define the Black-Scholes call option pricing formula
def black_scholes_call(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    return call_price

# Define the Delta of a call option
def delta_call(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    delta = norm.cdf(d1)
    return delta

# Calculate deltas and price differences for options with different volatilities
deltas = np.array([delta_call(S0, K, T, r, sigma) for K in Ks])

#Want to minimize the cost of our portfolio
c  = np.array([ black_scholes_call(S0, K, T, r, sigma) for K in Ks])

print(deltas, c)
# Constraints
# 1. Delta neutrality: The sum of deltas * number of options for each strike must be 0
A_eq = deltas.reshape(1, -1)
b_eq = [0]


x_bounds = [(-100, 100) for _ in Ks] 

# Solve the linear programming problem with cost constraint
result = linprog(c,  A_eq=A_eq, b_eq=b_eq, bounds=x_bounds, method='highs')


if result.success:
    # Extract the optimal number of options for each strike price
    optimal_quantities = result.x
    max_profit = -result.fun  # Negative because we minimized the negative profit
    print("Optimal number of options for each strike price:", optimal_quantities)
    print("Maximum profit:", max_profit)
else:
    print("Optimization was not successful. Error:", result.message)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the Black-Scholes call option pricing formula
def black_scholes_call(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    return call_price

# Define the Delta of a call option
def delta_call(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    delta = norm.cdf(d1)
    return delta

# Parameters
n_simulations = 100
n_periods = 1000  # Number of periods in each simulation
up_factor = 1.5  # Asset goes up by 50%
down_factor = 2/3  # Asset goes down by 33.33%%

# Function to simulate the process over time for visualization
def simulate_process_over_time(n_periods, up_factor, down_factor, n_simulations, balanced=True):
    all_paths = np.zeros((n_simulations, n_periods + 1))  # +1 to include the initial value
    for i in range(n_simulations):
        values = [1]  # Start each simulation with a value of 1
        stock = 0.5  # Initial stock proportion for balanced portfolio
        cash = 0.5  # Initial cash proportion for balanced portfolio
        for _ in range(n_periods):
            if np.random.rand() < 0.5:
                change_factor = up_factor
            else:
                change_factor = down_factor
            
            if balanced:
                # Update stock value and rebalance
                stock *= change_factor
                total = stock + cash
                stock = total / 2
                cash = total / 2
                values.append(stock + cash)
            else:
                # Update value without rebalancing
                values.append(values[-1] * change_factor)
                
        all_paths[i] = values

    return all_paths

# Generate paths
unbalanced_paths = simulate_process_over_time(n_periods, up_factor, down_factor, n_simulations, balanced=False)
balanced_paths = simulate_process_over_time(n_periods, up_factor, down_factor, n_simulations, balanced=True)

# Plotting
fig, axs = plt.subplots(2, 1, figsize=(10, 12))

# Unbalanced Portfolio Paths
for path in unbalanced_paths:
    axs[0].plot(path)
axs[0].set_title('Unbalanced Portfolio Value Over Time')
axs[0].set_xlabel('Time Period')
axs[0].set_ylabel('Portfolio Value')
axs[0].set_yscale('log')
# Balanced Portfolio Paths
for path in balanced_paths:
    axs[1].plot(path)
axs[1].set_title('Balanced Portfolio Value Over Time')
axs[1].set_xlabel('Time Period')
axs[1].set_ylabel('Portfolio Value')
axs[1].set_yscale('log')
plt.tight_layout()
plt.show()

```
